```python
import pandas as pd
import glob
import numpy as np
import os

from PIL import Image
import cv2
from skimage import color

import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# source code: https://www.kaggle.com/ratthachat/aptos-eye-preprocessing-in-diabetic-retinopathy
def crop_image_from_gray(img,tol=7):
    if img.ndim ==2:
        mask = img>tol
        return img[np.ix_(mask.any(1),mask.any(0))]
    elif img.ndim==3:
        gray_img = cv2.cvtColor(img, cv2.COLOR_RGB2GRAY)
        mask = gray_img>tol
        
        check_shape = img[:,:,0][np.ix_(mask.any(1),mask.any(0))].shape[0]
        if (check_shape == 0):
            return img
        else:
            img1=img[:,:,0][np.ix_(mask.any(1),mask.any(0))]
            img2=img[:,:,1][np.ix_(mask.any(1),mask.any(0))]
            img3=img[:,:,2][np.ix_(mask.any(1),mask.any(0))]
            img = np.stack([img1,img2,img3],axis=-1)
        return img
```

```python
# source code: https://www.kaggle.com/taindow/pre-processing-train-and-test-images
def circle_crop(img):   
    """Create circular crop around image center"""    
      
    height, width, depth = img.shape    
    
    x = int(width/2)
    y = int(height/2)
    r = np.amin((x,y))
    
    circle_img = np.zeros((height, width), np.uint8)
    cv2.circle(circle_img, (x,y), int(r), 1, thickness=-1)
    img = cv2.bitwise_and(img, img, mask=circle_img)
    img = crop_image_from_gray(img)
    
    return img 
```

```python
class ArrayFuncs:
    def __init__(self, array):
        self.array = array
        
    def crop_images(self):
        """Crops the images in the array to include only the fundus image"""
        cropped_image = [circle_crop(image) for image in self]
        return cropped_image

    def resize_images(self, dimension=224):
        """Converts the image dimensionality"""
        resized_image = [cv2.resize(image, dsize=(dimension, dimension),
                                    interpolation=cv2.INTER_CUBIC) for image in self]
        return resized_image

    def gray_filter(self):
        """Converts colored image into grayscale"""
        gray_image = [color.rgb2gray(image) for image in self]
        return gray_image

    def master_filter(self):
        """Converts colored image into a filtered image"""
        filtered_image = cv2.cvtColor(self,cv2.COLOR_BGR2RGB)
        filtered_image = cv2.addWeighted(filtered_image,4,cv2.GaussianBlur(filtered_image,(0,0),10),-4 ,128) 
        return filtered_image
```

```python
def plot_comparison(original, filtered, title_filtered):
    """Plots two images for comparison"""
    fig,(ax1, ax2) = plt.subplots(ncols=2, figsize=(40,40), sharex=True, sharey=True)
    ax1.imshow(original, cmap=plt.cm.gray)
    ax1.set_title('original')
    ax1.axis('off')
    ax2.imshow(filtered, cmap=plt.cm.gray)
    ax2.set_title(title_filtered)
    ax2.axis('off')
```

```python
# read the image data along with the path
image_list=[]
path_list=[]

for file in glob.glob('../data/MESSIDOR_images/Base*/*.tif'):
        im = Image.open(file)
        pl = Image.Path(file)
        image_list.append(im)
        path_list.append(pl)
        
assert (len(image_list) == len(path_list))
```

```python
# get the path name
file_path=[]

for path in path_list:
    file_path.append(path.name)
```

```python
# convert the images to a numpy array
array_images=[]
for file in image_list:
    array_images.append(np.array(file))

print(len(array_images))
print(array_images[0].shape)
```

```python
# crop the images
cropped_images = ArrayFuncs.crop_images(array_images)
print(cropped_images[0].shape)
```

```python
# resize the images
resized_images = ArrayFuncs.resize_images(cropped_images, dimension=224)
print(resized_images[0].shape)
```

```python
plot_comparison(resized_images[0], ArrayFuncs.master_filter(resized_images[0]), 'Filtered image')
```

```python
plot_comparison(resized_images[1], ArrayFuncs.master_filter(resized_images[1]), 'Filtered image')
```

```python
plot_comparison(resized_images[3], ArrayFuncs.master_filter(resized_images[3]), 'Filtered image')
```

```python
def default_function(image_array):
    img = image_array
    return img
```

```python
def save_images(name_series,output_folder,image_series, function=default_function, append_name=''): 
    """Saves image by diagnosis and applies a function and appends a name if stated"""
    input_name=name_series.values
    output_folder=output_folder
    
    indexer = 0
    
    for imagePath in input_name:
        
        input_array = image_series[indexer]
        
        filtered_image = function(input_array)
        
        img = Image.fromarray(filtered_image)
        
        fullOutPath = os.path.join(output_folder,append_name+imagePath) 
        
        img.save(fullOutPath)
        
        indexer += 1 
```
